```python
import numpy as np
import cv2

img = cv2.imread('image/house.png')
mask = np.zeros(img.shape, dtype=np.uint8)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret,thresh = cv2.threshold(gray,127,255,1)

contours,h = cv2.findContours(thresh,1,2)

for cnt in contours:
    approx = cv2.approxPolyDP(cnt,0.01*cv2.arcLength(cnt,True),True)
    print(len(approx))
    if len(approx)==5:
        print("pentagon")
        cv2.drawContours(mask,[cnt],0,255,-1)
    elif len(approx)==3:
        print("triangle")
        cv2.drawContours(mask,[cnt],0,(255,255,255),-1)
    elif len(approx)==4:
        print("square")
        cv2.drawContours(mask,[cnt],0,(255,255,255),-1)
    elif len(approx) == 9:
        print("half-circle")
        cv2.drawContours(mask,[cnt],0,(255,255,255),-1)
    elif len(approx) > 15:
        print("circle")
        cv2.drawContours(mask,[cnt],0,(255,255,255),-1)
#print(mask)
cv2.imshow('img',mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
#try for similar images and then take an average to obtain the saliency probability mask 
```
